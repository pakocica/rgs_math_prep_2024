# Solution

```python
# @title Problem 3

import sympy as sp

x, y = sp.symbols('x y')
F = (x-1)**2 + (x-y)**2 + 1
# F = 2*x**2 + 2*x*y + y**2 - 2*x + 2

F = F.expand()

print('a) FOC:')

FOC = {sp.diff(F,x), sp.diff(F,y)}

display(FOC)

sol = sp.solve(FOC, [x,y])

print('a) Solution:')
display(sol)

print('')

```

```python
# @title Plot of the Function $F(x,y)$

#@title Plot a 3D graph of the function $f(x,y) = x^2 + y^2$

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sympy as sp

# Define the symbolic variables
x, y = sp.symbols('x y')

def plot_3d_graph(f):
  # Convert the symbolic function to a numerical one (lambdify)
  f_lambdified = sp.lambdify((x, y), f, 'numpy')

  # Create a grid of points
  x_vals = np.linspace(-2, 4, 100)
  y_vals = np.linspace(-4, 4, 100)
  X, Y = np.meshgrid(x_vals, y_vals)
  Z = f_lambdified(X, Y)

  # Create a figure
  fig = plt.figure(figsize=(12, 6))

  # Plot the 3D surface
  ax1 = fig.add_subplot(121, projection='3d')
  ax1.plot_surface(X, Y, Z, cmap='plasma')

  # Plot the contour plot
  ax2 = fig.add_subplot(122)
  contour = ax2.contour(X, Y, Z, cmap='plasma')
  ax2.clabel(contour, inline=True, fontsize=8)

  # Add titles and labels
  ax1.set_title('Graph of function $' + sp.latex(f) + '$')
  ax1.set_xlabel('X')
  ax1.set_ylabel('Y')
  ax1.set_zlabel('Z')

  ax2.set_title('Contour (level curves)')
  ax2.set_xlabel('X')
  ax2.set_ylabel('Y')

  # Show the plots
  plt.show()

plot_3d_graph(F)
```
